````python

1.Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

##Code##

class DiscreteRandomVariable:
    def __init__(self, probabilities, values):
       
        if len(probabilities) != len(values):
            raise ValueError("Probabilities and values must have the same length.")
        if not (0.999 <= sum(probabilities) <= 1.001):
            raise ValueError("The sum of probabilities must be 1.")

        self.probabilities = probabilities
        self.values = values

    def expected_value(self):
       
        return sum(p * v for p, v in zip(self.probabilities, self.values))

    def variance(self):
    
        
        
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for p, v in zip(self.probabilities, self.values))

# Example usage
if __name__ == "__main__":
    probabilities = [0.2, 0.5, 0.3]
    values = [1, 2, 3]
    
    drv = DiscreteRandomVariable(probabilities, values)
    
    print("Expected Value:", drv.expected_value())
    print("Variance:", drv.variance())
````

````python
3.Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.


##Code##

import numpy as np

def generate_samples(distribution_type, params, num_samples):
    
    if distribution_type == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, num_samples)
    elif distribution_type == 'poisson':
        lambda_ = params
        samples = np.random.poisson(lambda_, num_samples)
    else:
        raise ValueError("Unsupported distribution type. Use 'binomial' or 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

# Example usage
if __name__ == "__main__":
    # Parameters for the distributions
    binomial_params = (10, 0.5)  
    poisson_params = 5          

    num_samples = 1000

    # analyze binomial samples
    binomial_mean, binomial_variance = generate_samples('binomial', binomial_params, num_samples)
    print(f"Binomial Distribution (n=10, p=0.5):")
    print(f"Mean: {binomial_mean}")
    print(f"Variance: {binomial_variance}")

   #analyze Poisson samples
    poisson_mean, poisson_variance = generate_samples('poisson', poisson_params, num_samples)
    print(f"\nPoisson Distribution (lambda=5):")
    print(f"Mean: {poisson_mean}")
    print(f"Variance: {poisson_variance}")
````

````python
4.Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, 
and standard deviation of the samples.


##Code##


import numpy as np

def generate_gaussian_samples(mean, std_dev, num_samples):
    
    # random samples from a Gaussian distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)
    
    # Compute statistics
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage
if __name__ == "__main__":
    
    mean = 0      
    std_dev = 1  
    num_samples = 1000  

    # samples and compute statistics
    mean_estimate, variance_estimate, std_dev_estimate = generate_gaussian_samples(mean, std_dev, num_samples)
    
    # Output the results
    print(f"Gaussian Distribution (mean={mean}, std_dev={std_dev}):")
    print(f"Estimated Mean: {mean_estimate}")
    print(f"Estimated Variance: {variance_estimate}")
    print(f"Estimated Standard Deviation: {std_dev_estimate}")
````

````python
5.Write a Python 6unction to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.

##Code##

import numpy as np

def normal_pdf(x, mean, std_dev):
    
    # the PDF using the formula
    coefficient = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = np.exp(- (x - mean)**2 / (2 * std_dev**2))
    pdf_value = coefficient * exponent
    
    return pdf_value

# Example usage
if __name__ == "__main__":
    # Parameters for the normal distribution
    mean = 0      # Mean of the distribution
    std_dev = 1   # Standard deviation of the distribution
    
    # Value for which to calculate the PDF
    x = 0
    
    # Calculate and print the PDF
    pdf_value = normal_pdf(x, mean, std_dev)
    print(f"PDF of normal distribution at x={x} (mean={mean}, std_dev={std_dev}): {pdf_value}")
````

````python
6.Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.


##Code##

import numpy as np

def exponential_cdf(x, rate):

    # CDF using the formula
    cdf_value = 1 - np.exp(-x / rate)
    
    return cdf_value

# Example usage
if __name__ == "__main__":
    # Parameters for the exponential distribution
    rate = 1.0  
    
    
    x_values = [0, 1, 2, 5, 10]
    
    # Calculate and print the CDF for each value
    for x in x_values:
        cdf_value = exponential_cdf(x, rate)
        print(f"CDF of exponential distribution at x={x} (rate={rate}): {cdf_value}")
````
